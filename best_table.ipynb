```python
from img2table.document import PDF
import os
import fitz

pdf_path = "Кинеф Потребность 2024.pdf"
pdf_doc = fitz.open(pdf_path)
pages_num = pdf_doc.page_count
```

```python
pages_num
```

```python
from img2table.ocr import SuryaOCR

ocr_surya = SuryaOCR(langs=["en", "ru"])
```

```python
from img2table.ocr import TesseractOCR

ocr_tess = TesseractOCR(lang='eng+rus')
```

```python
tables_per_page_df = []
for page_num in range(pages_num):
    pdf = PDF(src="Кинеф Потребность 2024.pdf", pages=[page_num])
    extracted_tables = pdf.extract_tables(ocr=ocr_tess,
                                      implicit_rows=False,
                                      borderless_tables=False,
                                      min_confidence=50)  # Экстракция таблиц
    df = extracted_tables[page_num][0].df
    df = df.rename(columns=df.iloc[0]).drop(df.index[0])
    tables_per_page_df.append(df)
```

```python
#удаление лишних столбцов 
for i in range(len(tables_per_page_df)):
    tables_per_page_df[i] = tables_per_page_df[i].loc[:, ~tables_per_page_df[i].columns.duplicated()]
```

```python
tables_per_page_df[0].fillna('')
```

```python
headers = tables_per_page_df[0].columns
for i in range(len(tables_per_page_df)):
    tables_per_page_df[i].columns = headers
    tables_per_page_df[i] = tables_per_page_df[i].fillna('')
```

```python

import re

def fix_date_suffix(text):
    return re.sub(r'(\d{1,2}\.\d{1,2}\.\d{4}|\d{4})\s?r\.', r'\1 г.', text)

for i in range(len(tables_per_page_df)):
    for head in tables_per_page_df[i].columns:
        tables_per_page_df[i][head] = tables_per_page_df[i][head].apply(fix_date_suffix)
```

```python
tables_per_page_df[0]
```

```python
import pandas as pd

merged_df = pd.concat(tables_per_page_df, axis=0)
merged_df = merged_df.reset_index(drop=True)
# Save the result to a new Excel file
merged_df.to_excel('merged_table.xlsx', index=False)
```

```python
import numpy as np
from openpyxl import load_workbook


file_name = 'merged_table.xlsx'
wb = load_workbook(file_name)
ws = wb.active

if ws.max_column>1:
    for indx, row in merged_df.iterrows():
        row_data =  pd.Series(row.to_list())
        if len(np.unique(row_data))==1:
            ws.merge_cells(start_row=indx+2, start_column=1, end_row=indx+2, end_column=ws.max_column)

wb.save(file_name)
```

```python
# Открываем существующий файл Excel
wb =load_workbook('merged_table.xlsx')
ws = wb.active  # выбираем активный лист

# Получаем максимальное количество строк и столбцов
max_row = ws.max_row
max_col = ws.max_column

# Проход по каждому столбцу
for col in ws.iter_cols(min_row=1, max_row=max_row, min_col=1, max_col=max_col):
    start_row = 1  # Начальная строка для объединения
    for row in range(2, max_row + 1):
        # Проверяем, равны ли текущая ячейка и предыдущая
        if col[row - 1].value != col[start_row - 1].value:
            # Если значение изменилось, объединяем ячейки, если диапазон больше 1
            if row - start_row > 1:
                ws.merge_cells(start_row=start_row, start_column=col[0].column, 
                               end_row=row - 1, end_column=col[0].column)
            start_row = row  # Обновляем начальную строку для нового диапазона
    # Объединяем последние ячейки в столбце, если они одинаковы
    if max_row - start_row > 0:
        ws.merge_cells(start_row=start_row, start_column=col[0].column, 
                       end_row=max_row, end_column=col[0].column)

# Сохраняем файл
wb.save('merged_table.xlsx')
```

```python
import numpy as np
def replace_same_values_with_nan(row):
    if row.nunique() == 1:  # Проверяем, все ли значения одинаковые
        return [np.nan] * len(row)  # Возвращаем строку с NaN
    return row  # Возвращаем исходную строку, если значения разные
# Удаляем строки, где все значения одинаковы
df_numeric= extracted_tables_nums[0][0].df[~extracted_tables_nums[0][0].df.apply(lambda row: row.nunique() == 1, axis=1)]

df_numeric_origin = extracted_tables_nums[0][0].df.apply(replace_same_values_with_nan, axis=1)
```

```python
#замена заголовков 
def change_headers(df):
    df.columns = df.iloc[0]
    df = df[1:].reset_index(drop=True)
    return df
df_numeric = change_headers(df_numeric)
df_numeric_origin = change_headers(df_numeric_origin)
```

```python
import pandas as pd


# Функция для проверки, что колонка содержит только числовые значения
def is_numeric_column(col):
    # Пробуем преобразовать к числовому типу
    if isinstance(col, pd.Series):
        # Заменяем запятую на точку и пытаемся преобразовать к числовому типу
        col = col.str.replace(',', '.', regex=False)  # Заменяем запятые на точки
        return pd.to_numeric(col, errors='coerce').notna().all()
    else:
        return False

# Получаем список колонок, в которых все данные числовые
numeric_columns = [col for col in df_numeric.columns if is_numeric_column(df_numeric[col])]

print(numeric_columns)
```

## postproccessing the table 

```python
import pandas as pd

# Чтение файла .xlsx и преобразование в DataFrame
df = pd.read_excel('tables_surya.xlsx')
```

```python
#улучшение числовых значений 
headers_origin  = np.array(df_numeric_origin.columns.tolist())
headers_table = np.array(df.columns.tolist())
for col in numeric_columns:
    indx = np.where(headers_origin == col)[0][0]
    df[headers_table[indx]] = df_numeric_origin[col]
```

```python
# Удаление колонок с названием "Unnamed"
df = df.drop(columns=[col for col in df.columns if 'Unnamed' in col])
```

```python
import re

# Функция для очистки от артефактов
def clean_data(value):
    if isinstance(value, str):
        # Оставляем только буквы, цифры и базовую пунктуацию
        return re.sub(r'[^A-Za-zА-Яа-я0-9.,:()\' ]+', '', value)
    return value

# Применение очистки ко всем ячейкам таблицы
df = df.applymap(clean_data)
df = df.fillna('')
```
